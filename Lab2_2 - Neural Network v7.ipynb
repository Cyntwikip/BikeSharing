## Implement NN

```python
from sklearn.metrics import r2_score

class NN():
    
    def __init__(self, layers, activations, gamma):
        self.layers = layers
        self.activations = activations
        self.gamma = gamma
        self.cache = {}
        
        np.random.seed(42)
        self.weights = {}
        w = 0.001
        #w = 1
        for i in range(1,len(self.layers)):
            self.weights['w'+str(i)] = w*(2*np.random.random((self.layers[i], self.layers[i-1]))-1)
            
        self.best_weights = self.weights
        self.best_epoch = 1
             
    def forward_propagation(self, X, weights):
        self.cache = {}
        layer_curr = X.T
        self.cache['a0'] = layer_curr
        
        for i in range(1, len(self.layers)):
            layer_pre = weights['w'+str(i)].dot(layer_curr)
            self.cache['a'+str(i)] = layer_pre
            layer_curr = self.activations[i](layer_pre)
            
        #print('Forward prop final shape: ', layer_curr.shape)
        return layer_curr
    
    def backward_propagation(self, error):
        layer_error = error
        for i in range(1, len(self.layers))[::-1]:
            layer_curr = self.cache['a'+str(i)]
            layer_pre = self.cache['a'+str(i-1)]
            if i != len(self.layers)-1:
                layer_error = self.weights['w'+str(i+1)].T.dot(layer_error)
            layer_error = layer_error*self.activations[i](layer_curr, deriv=True)
            self.weights['w'+str(i)] -= self.gamma * layer_error.dot(self.activations[i-1](layer_pre.T))
            
    def compute(self, X, y, X_test, y_test, iterations, metric, is_acc=True):
        y = y.reshape(1, -1)
        y_test = y_test.reshape(1, -1)
        
        error_history = []
        error_history_test = []
        accs_train = []
        accs_test = []
                
        for i in range(1, iterations+1):
            if i%500==0:
                print('Epoch {}'.format(i))
            
            weights = self.weights
            
            #test
            y_pred = self.forward_propagation(X_test, weights)
            y_pred = y_pred.reshape(1, -1)
            error = y_pred - y_test
            error_history_test.append(np.mean(abs(error), axis=1))
            
            #test accuracy
            acc = metric(y_pred, y_test)
            accs_test.append(acc)

            #train
            y_pred = self.forward_propagation(X, weights)
            y_pred = y_pred.reshape(1, -1)
            #print(y_pred)
            error = y_pred - y
            error_history.append(np.mean(abs(error), axis=1))
            
            #train accuracy
            acc = metric(y_pred, y)
            accs_train.append(acc)
            
            self.backward_propagation(error)
            
            if (i>1) and is_acc and (max(accs_test[:-1]) < accs_test[-1]):
                    self.best_epoch = i
                    self.best_weights = self.weights
            elif (i>1) and not is_acc and (min(accs_test[:-1]) > accs_test[-1]):
                    self.best_epoch = i
                    self.best_weights = self.weights
               
        if is_acc:
            self.best_acc = max(accs_test)
        else:
            self.best_acc = min(accs_test)
        self.error_hist_train = error_history 
        self.error_hist_test = error_history_test
        return
        # return error_history, error_history_test, accs_train, accs_test
```

```python
class Activation():
    def get(self, type):
        if type == 'linear':
              return self.linear
        elif type == 'sigmoid':
              return self.sigmoid
        elif type == 'sine':
              return self.sine
        elif type == 'tanh':
              return self.tanh
        elif type == 'relu':
              return self.relu
            
    @staticmethod
    def linear(x, deriv=False):
        if(deriv==True):
            return 1
        return x
            
    @staticmethod
    def sigmoid(x, deriv=False):
        sigmoid = 1/(1+np.exp(-x))
        if(deriv==True):
            return sigmoid*(1-sigmoid)
        return sigmoid
    
    @staticmethod
    def sine(x, deriv=False):
        if(deriv==True):
            return np.cos(x)
        return np.sin(x)
    
    @staticmethod
    def tanh(x, deriv=False):
        if(deriv==True):
            return 1-np.tanh(x)**2
        return np.tanh(x)
    
    @staticmethod
    def relu(x, deriv=False):
        if(deriv==True):
            return x>0
        return x*(x>0)
    
class EvaluationMetric():
    def get(self, type):
        if type == 'r-squared':
            metric = lambda ypred, ytrue: max(0,r2_score(ypred.flatten(), ytrue.flatten()))
        elif type == 'mape':
            metric = lambda ypred, ytrue: np.mean(np.abs((ytrue - ypred) / ytrue))
        elif type == 'rmlse':
            metric = lambda ypred, ytrue: np.sqrt((np.log((ypred + 1)/(ytrue + 1))**2).mean())
        
        return metric
            
```

## Using RMLSE

RMLSE cannot handle negative values. Thus, target feature must contain values from 0 to $+ \infty$.

```python
act = Activation()
# first activation must be linear
activations = [act.get('linear'),act.get('relu'),act.get('linear')]

# first hidden layer is a buffer layer.
nn = NN((56,28,1),activations,0.00001)

scalerY = MinMaxScaler()
y_train_scaled = scalerY.fit_transform(y_train)
y_test_scaled = scalerY.transform(y_test)
y_future_scaled = scalerY.transform(y_future)

metric = EvaluationMetric().get('rmlse')
nn.compute(X_train, y_train_scaled, X_test, y_test_scaled, 10000, metric, False)

print('----------')
print('Saving model')
pickle.dump(nn, open('pickle/model1.pickle', 'wb'))
```

## Using MAPE

Only use MAPE if the target feature only has positive values. We would not use any scaling here because the target feature already contains only positive values.

```python
act = Activation()
# first activation must be linear
activations = [act.get('linear'),act.get('relu'),act.get('linear')]

# first hidden layer is a buffer layer.
nn = NN((56,28,1),activations,0.00001)

scalerY = MinMaxScaler(feature_range=(1,2))
y_train_scaled = scalerY.fit_transform(y_train)
y_test_scaled = scalerY.transform(y_test)
y_future_scaled = scalerY.transform(y_future)

metric = EvaluationMetric().get('mape')
nn.compute(X_train, y_train_scaled, X_test, y_test_scaled, 10000, metric, False)

print('----------')
print('Saving model')
pickle.dump(nn, open('pickle/model2.pickle', 'wb'))
```

```python
scalerY = MinMaxScaler()
y_train_scaled = scalerY.fit_transform(y_train)
y_future_scaled = scalerY.transform(y_future)

metric = EvaluationMetric().get('rmlse')
nn = pickle.load(open('pickle/model1.pickle', 'rb'))

error_hist_train = nn.error_hist_train
error_hist_val = nn.error_hist_test
acc_val = nn.best_acc
epoch = nn.best_epoch

future_pred = nn.forward_propagation(X_future, nn.best_weights)
evaluation = metric(future_pred.reshape(1,-1), y_future_scaled.reshape(1,-1))
future_pred = scalerY.inverse_transform(future_pred)
```
